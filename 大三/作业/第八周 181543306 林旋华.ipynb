# 1. 加载数据集sns.load_dataset('flights')，该表包含年份、月份、乘客数，练习Seaborn的各种绘图命令。

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
flights=sns.load_dataset('flights')
flights.head()
```

```python
sns.scatterplot(x="year", y="passengers", data=flights)
```

```python
sns.lmplot(x="year", y="passengers", data=flights)
```

```python
sns.regplot(x="year", y="passengers", data=flights)
```

```python
sns.jointplot(x="year", y="passengers", data=flights)
```

```python
sns.residplot(x="year", y="passengers", data=flights)
```

```python
sns.lineplot(x="month", y="passengers", data=flights)
```

```python
# df.pivot用法
flights=flights.pivot('month','year','passengers') # 列索引，行索引 值
flights
```

```python
ax=sns.heatmap(flights)
```

```python
# 添加注释
ax=sns.heatmap(flights,annot=True,fmt='d')
```

```python
# 增加间距
x=sns.heatmap(flights,linewidths=0.5)
```

```python
ax=sns.heatmap(flights,cmap='YlGnBu') # Yl 黄 ，Gn 绿，Bu 蓝
```

```python
# 隐藏cbar
ax=sns.heatmap(flights,cmap='YlGnBu',cbar=False) # Yl 黄 ，Gn 绿，Bu 蓝
```
