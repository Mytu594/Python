```python
# 编写程序，先设置random.seed(7), 然后生成一个包含20个随机整数的列表，然后对其偶数下标的元素进行降序排列，奇数下标的元素不变
# 例如原始列表为  lst=[4,2,11,14,3,8], 排列后  lst=[3,2,4,14,11,8]
import random
random.seed(7)
lst=[random.randint(0,100) for _ in range(20)]
print('原列表',lst)

even=lst[::2]
odd=lst[1::2]
even.sort(reverse=True)

lst2=[x for yz in zip(even,odd) for x in yz]
print('新列表',lst2)
```

```python
# 编程，输出所有由1,2,3,4这4个数字组成的素数(1位/2位/3位/4位均可)，并且每个素数中每个数字只使用一次。
def iszhishu(x):
    if x==1:return False
    if x==2:return True
    for m in range(2, int(x**(1/2)+1)):
        if x%m==0:
            return False
    return True

from itertools import permutations #排列函数

dig=[]
for num in range(1,5):
    lst=list(permutations('1234',num))
    lst2=[int(''.join(yz))for yz in lst]
    dig.extend(lst2)
    
print('所有的排列情况：',dig)

lst=[x for x in dig if iszhishu(x)]
print('\n所有的质数为：',lst)
```

```python
# 编写函数 mysum , 可以对传入的任意单个数、元组或列表求和。
# 例如  mysum(1,2,(3,4),[5,6]) 则结果应为  21 
def mysum(*p):
    if len(p)==0:
        return 0
    else:
        if type(p[0])==range:
            return mysum(*list(p[0]))+mysum(*p[1:])
        elif type(p[0]) in (list,tuple):
            return mysum(*p[0])+mysum(*p[1:])
        else:
            return p[0] + mysum(*p[1:])

mysum(1,2,(3,4),[5,6])
```

```python
# 编写生成器函数 evenfun(m,n)，以生成器方式逐一返回任意范围 [m, n) 内的偶数。 m,n 是传入的参数。 
def evenfun(m,n):
    for x in range(m,n):
        if x%2==0:
            yield x

for y in evenfun(3,10):
    print(y)
```

```python
import jieba
from collections import Counter
with open('专业设置讨论.txt') as f:
    s = f.read()

lst = jieba.lcut(s)
lst = [x for x in lst if len(x)>0]
words = Counter(lst)

dic={m:n for m,n in words.most_common(100)}

from wordcloud import WordCloud
wc = WordCloud(font_path='simkai.ttf',background_color='white',width=500,height=400)
wc.generate_from_frequencies(dic).to_file('wc.png')
dic
```

```python
# 有英文文件  eng.txt , 要求将其每个单词都变为首字母大写，其它不变，存入 engnew.txt 中, 并分别统计输出每个字母开头的各有多少个单词。
# 例如，字母A开头有10个， 字母B开头有15, ... 字母Z开头 20个。
```

```python
with open('eng.txt') as f:
    s=f.read()
    
lst=s.split()
lst2=[x.capitalize() for x in lst]
news=''.join(lst2)

with open('engnew.txt','w')  as f:
    f.write(news)
    
lst3=[x[0] for x in lst2 if (x[0]>='A' and x[0]<='Z')]
from collections import Counter
dic=Counter(lst3)
print("统计结果",dic)
```

```python
# 有数据文件 score.csv , 要求输出【总评成绩】列的平均成绩，并输出总评成绩最高和最低的学号及总评成绩。
# 本例要求用python的原生文件和列表完成，不能使用pandas.

lst=[]
with open('score.csv',encoding='utf-8') as f:  # 保存为新文件
    s = f.readline()  # 先单独读一行，此为标题行
    print(s)
    for  s  in  f:
        stu=s.split(",")
        stu = stu[0:1] + [ float(x) for x in stu[1:]]
        lst.append(stu)

lst2 = sorted(lst, key=lambda x:x[4], reverse=True)
print('总评最高分 学号{} 成绩{}'.format(lst2[0][0], lst2[0][4]))
print('总评最低分 学号{} 成绩{}'.format(lst2[-1][0], lst2[-1][4]))

score=sum(x[4] for x in lst)   # 取每个学生的总评列 求和
print('总评平均分{:.1f}'.format(score/len(lst)))
```

```python
# 利用numpy的数组变换对图片 cat.jpg  完成下列操作
#(1) 在图片正中央标记贯穿整个图片的黑色的5个像素粗细的 十 字加号， 保存为 cat_x.jpg
#(2) 从上到下每10个像素高标记上布满整个图片的3像素粗细的白色横线， 保存为 cat_line.jpg
```

```python
import numpy as np
from PIL import Image
a=np.array(Image.open('cat.jpg'))
print(a.shape,a.dtype,a[0,0,:])

m=a.shape[0]
n=a.shape[1]

b=a.copy()
b[m//2-2:m//2+3]=0
b[:,n//2-2:n//2+3]=0

im=Image.fromarray(b)
im.save('cat_x.jpg')

b=a.copy()
b[::10]=255
b[1:10]=255
b[2:10]=255
im = Image.fromarray(b)
im.save('cat_line.jpg')
```
