```python
# 编写程序，先设置random.seed(7), 然后生成一个包含20个随机整数的列表，然后对其偶数下标的元素进行降序排列，奇数下标的元素不变
# 例如原始列表为  lst=[4,2,11,14,3,8], 排列后  lst=[11,2,4,14,3,8]
import random
random.seed(7)
lst=[random.randint(0,100) for _  in  range(20)]
print('原列表', lst)

even=lst[::2]      # 偶数索引
odd=lst[1::2]      # 奇数索引
even.sort(reverse=True)   # 偶数降序排列

lst2=[ x  for  yz  in  zip(even,odd)  for  x  in  yz ]   # 利用zip 将 偶奇 配对先得到元组 yz, 再对 yz 遍历 取出每个元素 x
print('新列表', lst2)
```

```python
# 放心，只是练习， 期末考试不会考这种

# 编程，输出所有由1,2,3,4这4个数字组成的质数(1位/2位/3位/4位均可)，并且每个质数中每个数字只使用一次。
# 按数学规定： 1和0即不是合数也不是质数 ,  2是质数

# 判断是否质数： 是返回True, 不是返回False
def  iszhishu(x):
    if x==1: return False
    if x==2: return True
    for  m  in  range(2, int(x**(1/2)+1)):
        if  x % m==0:
            return False
    return  True

from itertools import permutations   # permutations 是排列函数

dig=[]
for  num  in  range(1,5):
    lst=list(permutations('1234',num))    # 从"1234"中分别按 permutations('1234',1) .. permutations('1234',4) 取出 1/2/3/4 位的全部排列
    lst2=[int(''.join(yz)) for yz in lst]   # 初步结果可能为 yz=('1','2','3') , ''.join(yz) 连接为"123", 再 int() 转为整数 
    dig.extend(lst2)                 # 列表扩展

print('所有的排列情况:', dig)

lst=[x for x  in  dig if  iszhishu(x)]
print('\n所有的质数为:', lst)
```

```python
# 放心，只是练习， 期末考试不会考这种

# 编写函数 mysum , 可以对传入的任意单个数、元组或列表求和。
# 例如  mysum(1,2,(3,4),[5,6]) 则结果应为  21 

# 这个函数在 第5章 5.6视频中讲过 
def  mysum(*p):
    if  len(p) == 0:
        return 0
    else:
        if  type(p[0])== range:
            return mysum(*list(p[0])) + mysum(*p[1:])
        elif  type(p[0]) in (list,tuple):
            return mysum(*p[0]) + mysum(*p[1:])
        else:
            return p[0] + mysum(*p[1:])
        
mysum(1,2,(3,4),[5,6])
```

```python
# 编写生成器函数 evenfun(m,n)，以生成器方式逐一返回任意范围 [m, n) 内的偶数。 m,n 是传入的参数。 
def  evenfun(m,n):
    for  x  in  range(m,n):
        if x % 2==0:
            yield x           # 生成器函数的关键是使用 yield , 而不是普通的 return 
            
for  y  in  evenfun(3, 10):   # 测试
    print(y)
```

```python
# 以文件“专业设置讨论.txt”为素材，制作词云图。
import jieba
from collections import Counter         # 统计频次的函数
with open('专业设置讨论.txt') as f:     # 读取配套的文件
    s = f.read()
    
lst = jieba.lcut(s)                      # jieba分词得到词汇列表
lst = [x for x in lst if len(x)>1]      # 剔除长度为1的单字或标点符号
words = Counter(lst)                    # 统计频次，得到类似字典的统计结果

dic={m:n for m, n in words.most_common(100)}  # 利用排名前100位的词 生成频次字典

from   wordcloud  import  WordCloud
wc = WordCloud(font_path='simkai.ttf', background_color='white', width=500, height=400)  
wc.generate_from_frequencies(dic).to_file('wc.png')    # 本例利用  generate_from_frequencies 函数作图
dic
```

```python
# 有英文文件  eng.txt , 要求将其每个单词都变为首字母大写，其它不变，存入 engnew.txt 中, 并分别统计输出每个字母开头的各有多少个单词。
# 例如，字母A开头有10个， 字母B开头有15, ... 字母Z开头 20个。

with open('eng.txt') as f:           # 读取文件
    s = f.read()

lst=s.split()                       # 默认以空格分解，得到单词列表
lst2=[x.capitalize() for x in lst] # 每个单词转为首字母大写
news=' '.join(lst2)                 # 重新连接为 新的字符串

with open('engnew.txt','w') as f:  # 保存为新文件
    f.write(news)

lst3=[x[0] for x in lst2 if (x[0]>='A' and x[0]<='Z')]  # 取每个单词首字母。只保留首字符为字母的，这样剔除了标点符号和数字
from collections import Counter
dic=Counter(lst3)   # 统计结果
print("统计结果", dic)
```

```python
# 有数据文件 score.csv , 要求输出【总评成绩】列的平均成绩，并输出总评成绩最高和最低的学号及总评成绩。
# 本例要求用python的原生文件和列表完成，不能使用pandas.

lst=[]
with open('score.csv',encoding='utf-8') as f:  # 保存为新文件
    s = f.readline()  # 先单独读一行，此为标题行
    print(s)
    for  s  in  f:
        stu=s.split(",")
        stu = stu[0:1] + [ float(x) for x in stu[1:]]
        lst.append(stu)

lst2 = sorted(lst, key=lambda x:x[4], reverse=True)
print('总评最高分 学号{} 成绩{}'.format(lst2[0][0], lst2[0][4]))
print('总评最低分 学号{} 成绩{}'.format(lst2[-1][0], lst2[-1][4]))

score=sum(x[4] for x in lst)   # 取每个学生的总评列 求和
print('总评平均分{:.1f}'.format(score/len(lst)))
```
